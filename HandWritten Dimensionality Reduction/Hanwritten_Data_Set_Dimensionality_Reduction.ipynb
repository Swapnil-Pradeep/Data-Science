# Dimensionality Reduction

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.manifold import TSNE
import warnings
warnings.filterwarnings('ignore')
```

```python
df_tra=pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/optdigits/optdigits.tra',header=None)
df_tra.shape
```

```python
df_tes=pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/optdigits/optdigits.tes',header=None)
df_tes.shape
```

```python
df_tra.columns=['p'+str(i) for i in range(len(df_tra.columns)-1)]+['Y']
df_tes.columns=['p'+str(i) for i in range(len(df_tes.columns)-1)]+['Y']
```

```python
fig,ax=plt.subplots(1,20,figsize=(15,3),subplot_kw=dict(xticks=[],yticks=[]))
for i in np.arange(20):
  ax[i].imshow(df_tra.to_numpy()[i,0:-1].reshape(8,8),cmap=plt.cm.gray)
```

# PCA 

```python
df_tra=df_tra.loc[df_tra['Y'].isin([1,3,5])]
df_tes=df_tes.loc[df_tes['Y'].isin([1,3,5])]
```

```python
x_train=df_tra.drop('Y',axis=1)
y_train=df_tra['Y']
x_test=df_tes.drop('Y',axis=1)
y_test=df_tes['Y']
```

```python
pca=PCA()
trn_tra=pca.fit_transform(x_train)
plt.plot(pca.explained_variance_ratio_)
plt.show()
```

```python

sns.scatterplot(x=trn_tra[:,0],y=trn_tra[:,1],style=y_train,hue=y_train,palette=['red','blue','green'])
plt.show()
```

```python
pca.score(x_train)
```

```python
trn_tes=pca.transform(x_test)
sns.scatterplot(x=trn_tes[:,0],y=trn_tes[:,1],style=y_test,hue=y_test,palette=['red','blue','green'])
plt.show()
```

```python
pca.score(x_test)
```

# LDA

```python
lda= LinearDiscriminantAnalysis()
lda.fit(x_train,y_train)
trn_tra_1=lda.transform(x_train)
plt.plot(lda.explained_variance_ratio_)
plt.show()
```

```python

sns.scatterplot(x=trn_tra_1[:,0],y=trn_tra_1[:,1],style=y_train,hue=y_train,palette=['red','blue','green'])
plt.show()
```

```python
lda.score(x_train,y_train)
```
